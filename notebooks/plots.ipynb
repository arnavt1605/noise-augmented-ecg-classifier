```python
import matplotlib.pyplot as plt
import numpy as np

snr_levels = [0, 3, 6, 9, 12, 15, 18, 20]
cnn_baseline = [0.7123, 0.7678, 0.8493, 0.9038, 0.9429, 0.9686, 0.9834, 0.9891]
cnn_noiseaug = [0.8854, 0.9509, 0.9818, 0.9921, 0.9957, 0.9967, 0.9974, 0.9976]

plt.figure(figsize=(12,5))
plt.plot(snr_levels, np.array(cnn_baseline)*100, 'k--', label='Baseline CNN')
plt.plot(snr_levels, np.array(cnn_noiseaug)*100, 'b-', linewidth=2, label='CNN + Noise Augmentation')
plt.fill_between(snr_levels, np.array(cnn_baseline)*100, np.array(cnn_noiseaug)*100, color='skyblue', alpha=0.4, label='Improvement')
plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('Classification Accuracy (%)', fontsize=12)
plt.title('Performance vs SNR', fontsize=14)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('performance_vs_snr.png', dpi=300)
plt.show()
```
